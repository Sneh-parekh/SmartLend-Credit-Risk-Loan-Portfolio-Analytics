```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
from urllib.parse import quote_plus
from sqlalchemy import create_engine
```

```python
# Connect Python to MySQL

username = "root"
password = quote_plus("Karela@_2306")

engine = create_engine(
    f"mysql+pymysql://{username}:{password}@localhost/smartlend"
)
```

```python
customers = pd.read_sql(
            "SELECT * FROM customers",
            engine
)

customers.head()
```

```python
loans = pd.read_sql(
        " SELECT * FROM loans",
        engine
)

print(loans.shape)
```

```python
credit_risk = pd.read_sql(
    "SELECT * FROM credit_risk",
    engine
)

print(credit_risk.shape)
```

```python
df = customers.merge(
    loans,
    on="customer_id",
    how="inner"
)

df = df.merge(
    credit_risk,
    on="customer_id",
    how="inner"
)
```

```python
df.info()
```

```python
df['age'].describe()
```

```python
df['age_group'] = pd.cut(
    df['age'],
    bins=[18,25,35,45,55,100],
    labels=[
        '18-25',
        '26-35',
        '36-45',
        '46-55',
        '56+'
    ]
)
```

```python
df[['age','age_group']].head(10)
```

```python
df['age_group'].value_counts()
```

```python
df['annual_income'].describe()
```

```python
df['income_segment'] = pd.cut(
    df['annual_income'],
    bins= [0,1000000,2000000,3000000],
    labels= [
        'Low Income',
        'Middle Income',
        'High Income'
    ]
)
```

```python
df['income_segment'].value_counts()
```

```python
df['credit_score'].describe()
```

```python
df['risk_category'] = pd.cut(
    df['credit_score'],
    bins= [300,600,750,900],
    labels= [
        'High Risk',
        'Medium Risk',
        'Low Risk'
    ]
)
```

```python
df['risk_category'].value_counts()
```

```python
df['loan_term_months'].describe()
```

```python
df['loan_duration_bucket'] = pd.cut(
    df['loan_term_months'],
    bins= [0,24,60,120,300],
    labels= [
        'Short Term',
        'Medium Term',
        'Long Term',
        'Very Long Term'
    ]
)
```

```python
df['loan_duration_bucket'].value_counts()
```
